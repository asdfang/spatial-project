```python
import pandas as pd

f_age = "../data/agesex-IL-ziptab-2013-2017/ACS_17_5YR_S0101/ACS_17_5YR_S0101_with_ann.csv"

fr = pd.read_csv(f_age)

fr = fr.drop(0)
fr.head()
```

```python
chicago_zip_codes = [60290, 60601, 60602, 60603, 60604, 60605, 60606, 60607, 60608, 60610, 60611, 60614, 60615, 60618, 60619, 60622, 60623, 60624, 60628, 60609, 60612, 60613, 60616, 60617, 60620, 60621, 60625, 60626, 60629, 60630, 60632, 60636, 60637, 60631, 60633, 60634, 60638, 60641, 60642, 60643, 60646, 60647, 60652, 60653, 60656, 60660, 60661, 60664, 60639, 60640, 60644, 60645, 60649, 60651, 60654, 60655, 60657, 60659, 60666, 60668, 60673, 60677, 60669, 60670, 60674, 60675, 60678, 60680, 60681, 60682, 60686, 60687, 60688, 60689, 60694, 60695, 60697, 60699, 60684, 60685, 60690, 60691, 60693, 60696, 60701]
chicago_fr = fr.loc[fr['GEO.id2'].astype('int32').isin(chicago_zip_codes)]

age_ests = chicago_fr

age_ests.index = age_ests['GEO.id2']
age_ests = age_ests[['HC01_EST_VC01', 'HC01_EST_VC03', 'HC01_EST_VC04', 'HC01_EST_VC05', 'HC01_EST_VC06', 'HC01_EST_VC07', 'HC01_EST_VC08', 'HC01_EST_VC09', 'HC01_EST_VC10', 'HC01_EST_VC11', 'HC01_EST_VC12', 'HC01_EST_VC13', 'HC01_EST_VC14', 'HC01_EST_VC15', 'HC01_EST_VC16', 'HC01_EST_VC17', 'HC01_EST_VC18', 'HC01_EST_VC19', 'HC01_EST_VC20']]

age_ests = age_ests.astype('float64')

age_ests.head()

```

```python
def simpson_index(total, cols):
    s = sum([col*(col-1.0) for col in cols])
    return 1.0 - (s/(total * (total - 1.0)))

age_ests["Age Index"] = age_ests.apply(lambda row: simpson_index(row["HC01_EST_VC01"], [row["HC01_EST_VC03"],
                                                             row["HC01_EST_VC04"], row["HC01_EST_VC05"],
                                                             row["HC01_EST_VC06"], row["HC01_EST_VC07"],
                                                             row["HC01_EST_VC08"], row["HC01_EST_VC09"],
                                                             row["HC01_EST_VC10"], row["HC01_EST_VC11"],
                                                             row["HC01_EST_VC12"], row["HC01_EST_VC13"],
                                                             row["HC01_EST_VC14"], row["HC01_EST_VC15"],
                                                             row["HC01_EST_VC16"], row["HC01_EST_VC17"],
                                                             row["HC01_EST_VC18"], row["HC01_EST_VC19"],
                                                             row["HC01_EST_VC20"]]), axis=1)

age_ests = age_ests.rename(index=str, columns={"HC01_EST_VC01": "Total",
                                                 "HC01_EST_VC03": "< 5 yrs",
                                                 "HC01_EST_VC04": "5-9 yrs",
                                                 "HC01_EST_VC05": "10-14 yrs",
                                                 "HC01_EST_VC06": "15-19 yrs",
                                                 "HC01_EST_VC07": "20-24 yrs",
                                                 "HC01_EST_VC08": "25-29 yrs",
                                                 "HC01_EST_VC09": "30-34 yrs",
                                                 "HC01_EST_VC10": "35-39 yrs",
                                                 "HC01_EST_VC11": "40-44 yrs",
                                                 "HC01_EST_VC12": "45-49 yrs",
                                                 "HC01_EST_VC13": "50-54 yrs",
                                                 "HC01_EST_VC14": "55-59 yrs",
                                                 "HC01_EST_VC15": "60-64 yrs",
                                                 "HC01_EST_VC16": "65-69 yrs",
                                                 "HC01_EST_VC17": "70-74 yrs",
                                                 "HC01_EST_VC18": "75-79 yrs",
                                                 "HC01_EST_VC19": "80-84 yrs",
                                                 "HC01_EST_VC20": "85+ yrs",})

age_ests.head()
```
