```python
import pandas as pd
import numpy as np

f_income = "../data/income-IL-ziptab-2013-2017/ACS_17_5YR_S1901/ACS_17_5YR_S1901_with_ann.csv"

fr = pd.read_csv(f_income)

fr = fr.drop(0)
fr.head()
```

```python
chicago_zip_codes = [60290, 60601, 60602, 60603, 60604, 60605, 60606, 60607, 60608, 60610, 60611, 60614, 60615, 60618, 60619, 60622, 60623, 60624, 60628, 60609, 60612, 60613, 60616, 60617, 60620, 60621, 60625, 60626, 60629, 60630, 60632, 60636, 60637, 60631, 60633, 60634, 60638, 60641, 60642, 60643, 60646, 60647, 60652, 60653, 60656, 60660, 60661, 60664, 60639, 60640, 60644, 60645, 60649, 60651, 60654, 60655, 60657, 60659, 60666, 60668, 60673, 60677, 60669, 60670, 60674, 60675, 60678, 60680, 60681, 60682, 60686, 60687, 60688, 60689, 60694, 60695, 60697, 60699, 60684, 60685, 60690, 60691, 60693, 60696, 60701]
chicago_fr = fr.loc[fr['GEO.id2'].astype('int32').isin(chicago_zip_codes)]

income_ests = chicago_fr

income_ests.index = income_ests['GEO.id2']
income_ests = income_ests[['HC01_EST_VC01', 'HC01_EST_VC02', 'HC01_EST_VC03', 'HC01_EST_VC04', 'HC01_EST_VC05', 'HC01_EST_VC06', 'HC01_EST_VC07', 'HC01_EST_VC08', 'HC01_EST_VC09', 'HC01_EST_VC10', 'HC01_EST_VC11']]

income_ests = income_ests.astype('float64')
income_ests["HC01_EST_VC02"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC02"] / 100.0)
income_ests["HC01_EST_VC03"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC03"] / 100.0)
income_ests["HC01_EST_VC04"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC04"] / 100.0)
income_ests["HC01_EST_VC05"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC05"] / 100.0)
income_ests["HC01_EST_VC06"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC06"] / 100.0)
income_ests["HC01_EST_VC07"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC07"] / 100.0)
income_ests["HC01_EST_VC08"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC08"] / 100.0)
income_ests["HC01_EST_VC09"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC09"] / 100.0)
income_ests["HC01_EST_VC10"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC10"] / 100.0)
income_ests["HC01_EST_VC11"] = np.round(income_ests["HC01_EST_VC01"] * income_ests["HC01_EST_VC11"] / 100.0)


income_ests.head()
```

```python
def simpson_index(total, cols):
    s = sum([col*(col-1.0) for col in cols])
    return 1.0 - (s/(total * (total - 1.0)))

income_ests["Inc. Index"] = income_ests.apply(lambda row: simpson_index(row["HC01_EST_VC01"], [row["HC01_EST_VC02"], row["HC01_EST_VC03"],
                                                             row["HC01_EST_VC04"], row["HC01_EST_VC05"],
                                                             row["HC01_EST_VC06"], row["HC01_EST_VC07"],
                                                             row["HC01_EST_VC08"], row["HC01_EST_VC09"],
                                                             row["HC01_EST_VC10"], row["HC01_EST_VC11"]]), axis=1)

income_ests = income_ests.rename(index=str, columns={"HC01_EST_VC01": "Total",
                                                 "HC01_EST_VC02": "< $10k",
                                                 "HC01_EST_VC03": "[$10k, $15k)",
                                                 "HC01_EST_VC04": "[$15k, $25k)",
                                                 "HC01_EST_VC05": "[$25k, $35k)",
                                                 "HC01_EST_VC06": "[$35k, $50k)",
                                                 "HC01_EST_VC07": "[$50k, $75k)",
                                                 "HC01_EST_VC08": "[$75k, $100k)",
                                                 "HC01_EST_VC09": "[$100k, $150k)",
                                                 "HC01_EST_VC10": "[$150k, $200k)",
                                                 "HC01_EST_VC11": "$200k +",})

income_ests.head()
```

```python
output = income_ests[['Inc. Index']]

output.to_csv('income_diversity.csv')
output
```
