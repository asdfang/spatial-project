```python
import pandas as pd
```

```python
housing_df = pd.read_csv('data/housing/housing.csv', na_values='-')
housing_df = housing_df.drop(0)
housing_df = housing_df.drop(1)
# housing_df.head()

economic_df = pd.read_csv('data/economic/economic.csv', na_values=['-', '(X)', '**'])
economic_df = economic_df.drop(0)
economic_df = economic_df.drop(1)
economic_df.head()
```

```python
chicago_zip_codes = [60290, 60601, 60602, 60603, 60604, 60605, 60606, 60607, 60608, 60610, 60611, 60614, 60615, 60618, 60619, 60622, 60623, 60624, 60628, 60609, 60612, 60613, 60616, 60617, 60620, 60621, 60625, 60626, 60629, 60630, 60632, 60636, 60637, 60631, 60633, 60634, 60638, 60641, 60642, 60643, 60646, 60647, 60652, 60653, 60656, 60660, 60661, 60664, 60639, 60640, 60644, 60645, 60649, 60651, 60654, 60655, 60657, 60659, 60666, 60668, 60673, 60677, 60669, 60670, 60674, 60675, 60678, 60680, 60681, 60682, 60686, 60687, 60688, 60689, 60694, 60695, 60697, 60699, 60684, 60685, 60690, 60691, 60693, 60696, 60701]
chicago_housing_df = housing_df.loc[housing_df['GEO.id2'].astype('int32').isin(chicago_zip_codes)]
chicago_housing_df

chicago_economic_df = economic_df.loc[housing_df['GEO.id2'].astype('int32').isin(chicago_zip_codes)]
chicago_economic_df
```

```python
new_chicago_housing_df = chicago_housing_df 
new_chicago_housing_df.index = new_chicago_housing_df['GEO.id2']
new_chicago_housing_df = new_chicago_housing_df[['HC01_VC128', 'HC01_VC155', 'HC01_VC191']]
new_chicago_housing_df = new_chicago_housing_df.rename(index=str, columns={'HC01_VC128': 'median_housing_value',
                                                       'HC01_VC155': 'median_monthly_housing_cost',
                                                       'HC01_VC191': 'gross_rent',
                                                      })
new_chicago_housing_df = new_chicago_housing_df.replace('1,500+', 1500)
new_chicago_housing_df = new_chicago_housing_df.astype('float64')
new_chicago_housing_df

new_chicago_economic_df = chicago_economic_df 
new_chicago_economic_df.index = new_chicago_economic_df['GEO.id2']
new_chicago_economic_df = new_chicago_economic_df[['HC03_VC12', 'HC01_VC85', 'HC03_VC31', 'HC01_VC36']]
new_chicago_economic_df = new_chicago_economic_df.rename(index=str, columns={'HC03_VC12': 'unemployment_rate',
                                                       'HC01_VC85': 'median_household_income',
                                                       'HC03_VC31': 'walked_to_work_pct',
                                                       'HC01_VC36': 'mean_travel_time_to_work'
                                                      })
# new_chicago_economic_df = new_chicago_economic_df.replace('1,500+', 1500)
# new_chicago_economic_df = new_chicago_economic_df.astype('float64')
```

```python
new_df = pd.merge(new_chicago_housing_df, new_chicago_economic_df, on='GEO.id2')
```
